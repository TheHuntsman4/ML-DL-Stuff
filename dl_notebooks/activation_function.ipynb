```python
# Defining basic imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the range and defining the activation functions
x = np.linspace(-10, 10, 100)

relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
leaky_relu = np.where(x > 0, x, 0.01 * x)
```

```python
plt.plot(x, relu, label='ReLU')
plt.grid(True)
plt.title("ReLU")
plt.show()
```

```python
plt.plot(x, sigmoid, label='Sigmoid')
plt.grid(True)
plt.title("Sigmoid")
plt.show
```

```python
plt.plot(x, tanh, label='tanh')
plt.grid(True)
plt.title("tanh")
plt.show
```

```python
plt.plot(x, leaky_relu, label='Leaky ReLU')
plt.grid(True)
plt.title("Leaky ReLU")
plt.show
```

```python
plt.plot(x, relu, label='ReLU')
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, leaky_relu, label='Leaky ReLU')
plt.legend()
plt.grid(True)
plt.title("Activation Functions")
plt.show()
```
